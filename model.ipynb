```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
import pickle
from preprocessing import preprocess
```

```python
preprocessed_data=pd.read_csv("preprocessed.csv")
preprocessed_data
```

```python
x=preprocessed_data.drop("Sentiment",axis=1)
y=preprocessed_data["Sentiment"].map({j:i+1 for i,j in  enumerate(set(preprocessed_data["Sentiment"]))})
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

## Random Forest

```python
RandomF=RandomForestClassifier(n_estimators=100,random_state=42,criterion='entropy')
RandomF.fit(x_train,y_train)
```

```python
RandomFHistory=RandomF.predict(x_test)
```

```python
accuracy_score(y_test,RandomFHistory)
```

```python
with open("token_model.pkl","rb") as f:
    token_mod=pickle.load(f)
```

```python
def SentimentAnalysis(comment):
    preprocessed_comment = preprocess(comment)
    new_comment = token_mod.transform([preprocessed_comment]) 
    return new_comment
```

```python
class_labels = ['Positive', 'Neutral', 'Irrelevant', 'Negative']
prediction = RandomF.predict(SentimentAnalysis("@realdonuldtrump very shitty"))
mapped_prediction = class_labels[prediction[0]-1]
print(mapped_prediction)
```
